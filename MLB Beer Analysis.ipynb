```python
import pandas as pd
```

```python
# load data 
df = pd.read_csv('MLB Beer Prices.csv')
print(df.shape)
df.head()
```

```python
# statistics
df.describe()
```

```python
# teams in each year
df['Year'].value_counts()
```

```python
# average across the league by year
df.groupby('Year').mean()
```

```python
# team with the highest price per ounce of beer
df[df[' Price per Ounce '] == df[' Price per Ounce '].max()]
```

```python
# 2018 beer prices by team sorted highest to lowest
# New York doesn't make sense based on memory, but could be accurate
# estimating beers at Yankee Stadium in 2019 were somewhere around .58 Per Ounce
df[df['Year'] == 2018].sort_values(by=' Price per Ounce ', ascending=False)
```
